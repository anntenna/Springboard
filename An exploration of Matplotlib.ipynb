# An exploration of Matplotlib

This is a simple follow along of [J.R. Johansson](http://github.com/jrjohansson/scientific-python-lectures)'s lectures on Matplotlib.

```python
# This line configures matplotlib to show figures embedded in the notebook
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Using the MATLAB compatible API:
from pylab import *
```

```python
x = np.linspace(0,5,10)
y = x ** 2
```

```python
figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
#Subplot and color/symbol selection
subplot(1,2,1)
plot(x,y,'r--')
subplot(1,2,2)
plot(y,x,'g*-')
```

```python
#Using the object-oriented API
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
#Adding more than one axis to the figure
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

#main figure
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

#inset figure
axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('inset')
```

```python
#Using the subplots layout manager

fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
for axis in axes:
    axis.plot(x,y,'r')
    axis.set_xlabel('x')
    axis.set_ylabel('y')
    axis.set_title('title')
```

```python
#Using tight layout to avoid overlapping content
fig, axes = plt.subplots(nrows = 1, ncols = 2)
for axis in axes:
    axis.plot(x,y,'r')
    axis.set_xlabel('x')
    axis.set_ylabel('y')
    axis.set_title('title')
    
fig.tight_layout()
```

```python
#Setting figure size, aspect ratio and DPI

fig = plt.figure(figsize=(4,3), dpi = 100)
```

```python
fig, axes = plt.subplots(figsize=(12,4), dpi = 300)

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
#Saving an img
fig.savefig("graph.png")
```

```python
#Labeling curves correctly

x = np.linspace(0,5,10)
y = x ** 2
z = x ** 3

fig, axes = plt.subplots(figsize=(9,3), dpi = 200)

axes.plot(x, x, label='self')
axes.plot(x, y, label='squared')
axes.plot(x, z, label='cubed')
axes.legend(loc=9)
```

```python
#Using Matplotlib's support for LaTeX formatted text

x = np.linspace(0,5,10)
y = x ** 2
z = x ** 3

fig, axes = plt.subplots(figsize=(9,3), dpi = 200)

axes.plot(x, x, label=r'$y = \alpha$')
axes.plot(x, y, label=r'$y = \alpha^2$')
axes.plot(x, z, label=r'$y = \alpha^3$')
axes.legend(loc=2)

axes.set_xlabel(r'$\alpha$')
axes.set_ylabel(r'$y$')
```

```python
#Updating global matplotlib configuration parameters (font-size and font-family):

matplotlib.rcParams.update({'font.size':9, 'font.family':'serif'})
```

```python
x = np.linspace(0,5,10)
y = x ** 2
z = x ** 3

fig, axes = plt.subplots(figsize=(9,3), dpi = 200)

axes.plot(x, x, label=r'$y = \alpha$')
axes.plot(x, y, label=r'$y = \alpha^2$')
axes.plot(x, z, label=r'$y = \alpha^3$')
axes.legend(loc=2)

axes.set_xlabel(r'$\alpha$')
axes.set_ylabel(r'$y$') 
```

```python
#Different ways of defining colors and plot types, line width and line style, markers
fig, axes = plt.subplots()

axes.plot(x, x**2, 'b.-', alpha=0.7)
axes.plot(x, x**3, color='#DD11FF')
#possible linestyle options: -, --, -., :, steps 
axes.plot(x, x**4, color='green', linewidth=0.5, linestyle='steps')
axes.plot(x, x**3.5, color='orange', lw=1.5, marker='*')
plt.show()
```

```python
#Having some fun with the options

linestyles = ['-', '--', '-.', ':', 'steps']
markers = ['.', 'o', '*', 's', ',', '+', '1', '2', '3', '4']
linecolors = ['blue','#0044EE', 'red','magenta','#001335']
markerfacecolors = ['yellow','green','pink','white','#777777']
markeredgecolors = ['purple','indigo','violet','brown','black']

fig, axes = plt.subplots(figsize=(9,3), dpi = 200)

for i in range(1,6):
    axes.plot(x, 
              x**i, 
              color=linecolors[i-1], 
              linewidth=i**(-1), 
              marker=markers[i-1], 
              markerfacecolor=markerfacecolors[i-1],
              markeredgecolor=markeredgecolors[i-1],
              label = r'$y = \alpha^{}$'.format(i))
    axes.set_ylabel(r'$y$')
    axes.set_xlabel(r'$\alpha$')
    
axes.legend()
plt.show()
    
```
