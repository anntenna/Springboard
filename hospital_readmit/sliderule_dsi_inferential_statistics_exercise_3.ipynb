### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### First Impressions

My first impression of the preliminary report is that the conclusions are being drawn without performing due diligence and applying statistical rigor to the analysis of the hospital readmission rate data. I also disagree with using just the measure of the number of discharges to explain the readmission rate of hospitals. 

To quickly measure how much the two variables we're observing are correlated, let's find the **R-squared value** (correlation coefficient).

```python
import math
# Your turn
num_discharges = clean_hospital_read_df['Number of Discharges']
excess_readmit_rate = clean_hospital_read_df['Excess Readmission Ratio']
math.pow(excess_readmit_rate.corr(num_discharges),2)
```

This means that there is less than 1% correlation between the two variables, i.e., less than 1% of the variance in excess readmission rate is being explained by the number of discharges of a hospital. 

To further understand the statistical difference between hospitals that exceed the average readmission rate (excess readmission rate <= 1) and those that do not exceed the average readmission rate (excess readmission rate > 1), let's divide the data into two samples. In this case, we are trying to validate the choice of discharge numbers as predictor variable of excess readmission ratio, by using the excess readmission ratio as a "treatment" to divide the data into two independent samples.

```python
less_than_avg = clean_hospital_read_df[clean_hospital_read_df['Excess Readmission Ratio'] <= 1]['Number of Discharges']
more_than_avg = clean_hospital_read_df[clean_hospital_read_df['Excess Readmission Ratio'] > 1]['Number of Discharges']

n_less = len(less_than_avg)
n_more = len(more_than_avg)

mean_discharges_less = less_than_avg.mean()
mean_discharges_more = more_than_avg.mean()

std_discharges_less = less_than_avg.std()
std_discharges_more = more_than_avg.std()


print ('# discharges for hospitals with less than avg ERR: (Mean = {}. St dev = {})'.format(mean_discharges_less,std_discharges_less))
print ('# discharges for hospitals with more than avg ERR: (Mean = {}. St dev = {})'.format(mean_discharges_more,std_discharges_more))
print ('Num samples: less than avg ERR = {}, more than avg ERR = {}'.format(n_less, n_more))
```

```python
#Are the two samples normal?
import seaborn as sns
sns.distplot(less_than_avg,bins=1000)
```

```python
sns.distplot(more_than_avg,bins=1000)
```

### Choosing a hypothesis test

In order to compare these two samples, I'm going to pick the **two-sample t-test** because:
 - the sample sizes are large enough
 - the samples are not normal, but it doesn't matter because the sample sizes are large enough to be good estimates of the population
 - the standard deviations (variances) of the two samples are close enough to be replaced by a pooled variance while calculating the t-statistic

```python
from scipy import stats
np.set_printoptions(formatter={'float_kind':'{:f}'.format})

x = stats.ttest_ind(less_than_avg, more_than_avg, equal_var=False)
print (x)
print (['{:f}'.format(item) for item in x])
```

```python
df = n_less + n_more - 2
stats.t.isf(0.005,df)
```

As per the result of the t-test, the high t-statistic of 5.3 (far higher than t-critical value of 2.57) with a p-value close to zero (far less than our $\alpha$ value of 0.01) tells us that **the difference between the number of discharges of the two samples is statistically significant**. This wasn't immediately apparent from the descriptive statistics or the scatter plot in the preliminary assessment.

### Effect of Number of Discharges on Excess Readmission Ratio

To further understand the effect of the number of discharges on our dependent variable (ERR), let's divide up the data once again into two samples based on the median number of discharges of the data.


```python
median_num_discharges = clean_hospital_read_df['Number of Discharges'].median()
print (median_num_discharges)
```

```python
less_discharges = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] <= median_num_discharges]['Excess Readmission Ratio'].dropna()
more_discharges = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] > median_num_discharges]['Excess Readmission Ratio'].dropna()

mean_ERR_less = less_discharges.mean()
mean_ERR_more = more_discharges.mean()

std_ERR_less = less_than_avg.std()
std_ERR_more = more_than_avg.std()

n_less = len(less_discharges)
n_more = len(more_discharges)

print ('ER Ratio for hospitals with less than avg ERR: (Mean = {}. St dev = {})'.format(mean_ERR_less, std_ERR_less))
print ('ER Ratio for hospitals with more than avg ERR: (Mean = {}. St dev = {})'.format(mean_ERR_more, std_ERR_more))
print ('Num samples: less than median discharges = {}, more than median discharges = {}'.format(n_less, n_more))
```

```python
# Are the two samples normal?

sns.distplot(less_discharges, label='Lesser discharges')
sns.distplot(more_discharges, label='More discharges')
plt.legend()
pass
```

The distributions of ERR's seem to be normal. I'll choose the same test as above for this pair of samples as well.


```python
x = stats.ttest_ind(less_discharges,more_discharges, equal_var=False)
print (x)
print (['{:f}'.format(item) for item in x])
```

I'm going to verify this manually using the t-statistic formula for two-tailed t-test:
$$t-statistic = \frac{\bar{x_1} - \bar{x_2}}{\sqrt{\frac{S_p^2}{n_1} + \frac{S_p^2}{n_2}}}$$

```python
# Calculate sum of square differences for pooled variance
ss_less = less_discharges.var()*n_less
ss_more = more_discharges.var()*n_more

# Calculate pooled variance
pooled_var = (ss_less + ss_more)/(n_less + n_more -2)

# Calculate standard error
standard_error = math.sqrt(pooled_var/n_less + pooled_var/n_more)

# Calculate t-statistic
t_stat = (mean_ERR_less - mean_ERR_more)/standard_error

print (t_stat)
```

Again, we get a remarkably high t-statistic of 8.04 with a p-value close to zero. This means that the difference between the two means of excess readmission ratios is statistically significant.

### Statistical v/s Practical Significance

The statistical significance of the differences of the two samples we saw indicates that the two distributions are very different, but whether or not this difference matters depends highly on several other factors that we have not yet considered, such as the locations of the hospitals in the two samples. For example, hospitals in bigger cities might experience lower readmission rates because patients have more choices, whereas hospitals in smaller towns and villages would often be the only choice available for patients in the areas served by those hospitals.
